```python
temp = pd.read_csv('INMET_SE_MG_F501_BELO HORIZONTE - CERCADINHO_01-01-2024_A_30-04-2024.csv', encoding = 'ISO-8859-1', skiprows = 8, delimiter = ';', index_col = 'Data')
temp.head()
```

```python
#temp = temp[["TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)"]]
#print(temp)
temp = temp.iloc [:,[0,1,5,8,9,14,15,16]]

```

```python
temp.head()
```

```python
temp['Hora UTC'] = pd.to_datetime(temp['Hora UTC'], format='%H%M UTC')

# Criar uma nova coluna para representar os períodos do dia
def periodo_do_dia(hora):
    if hora.hour < 6:
        return '00:00 às 06:00'
    elif hora.hour < 12:
        return '06:00 às 12:00'
    elif hora.hour < 18:
        return '12:00 às 18:00'
    else:
        return '18:00 às 24:00'

temp['Periodo do Dia'] = temp['Hora UTC'].apply(periodo_do_dia)

# Agrupar os dados por períodos do dia e calcular a média
media_por_periodo = temp.groupby('Periodo do Dia').mean()

media_por_periodo.head()
```
